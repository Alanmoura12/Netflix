```python
import pandas as pd
```

```python
df_receita = pd.read_csv('NetflixsRevenue2018toQ2_2020.csv')
df_assinantes = pd.read_csv('CNetflixSubscribersbyCountryfrom2018toQ2_2020.csv')
```

```python
df_receita_long = df_receita.melt(id_vars=['Area'], var_name='Trimestre', value_name='Receita')
df_assinantes_long = df_assinantes.melt(id_vars=['Area'], var_name='Trimestre', value_name='Assinantes')
```

```python
df_consolidado = pd.merge(df_receita_long, df_assinantes_long, on=['Area', 'Trimestre'])
```

```python
df_consolidado['receita_por_assinante'] = df_consolidado['Receita'] / df_consolidado['Assinantes']
```

```python
df_consolidado['receita_por_assinante'] = df_consolidado['receita_por_assinante'].round(2)
```
